```python
# Python program implementing Image Steganography 
  
# PIL module is used to extract 
# pixels of image and modify it 
from PIL import Image 

# Convert encoding data into 8-bit binary 
# form using ASCII value of characters 
def genData(data): 
          
        # list of binary codes 
        # of given data 
        newd = []  
          
        for i in data: 
            newd.append(format(ord(i), '08b')) 
        return newd 

# Pixels are modified according to the 
# 8-bit binary data and finally returned 
def modPixl(pix, data, pixel_flag): 
      
    datalist = genData(data) 
    lendata = len(datalist) 
    imdata = iter(pix)
    imdata_copy = iter(pix)
  
    for i in range(lendata): 
        
        # Extracting 3 pixels at a time 
        pix = [value for value in imdata.__next__()[:3] +
                                  imdata.__next__()[:3] +
                                  imdata.__next__()[:3]]
                                      
        # Pixel value should be made  
        # odd for 1 and even for 0 
        for j in range(0, 8): 
            if (datalist[i][j]=='0') and (pix[j]% 2 != 0): 
                if(pixel_flag == 0):
                    zeroes = 0
                    ones = 0
                    for value in range(3):
                        if(imdata.__next__()[0] == 0):
                            zeroes += 1
                        elif(imdata.__next__()[0] == 1):
                            ones += 1
                    if(zeroes > ones):
                        if (pix[j]% 2 != 0): 
                            pix[j] -= 1
                    else:
                        continue
                
                elif(pixel_flag == 1):
                    zeroes = 0
                    ones = 0
                    for value in range(3):
                        if(imdata.__next__()[0] == 0):
                            zeroes += 1
                        elif(imdata.__next__()[0] == 1):
                            ones += 1
                    if(zeroes < ones):
                        if (pix[j]% 2 != 0): 
                            pix[j] -= 1
                    else:
                        continue
                    
                    
                      
            elif (datalist[i][j] == '1') and (pix[j] % 2 == 0): 
                 if(pixel_flag == 0):
                    zeroes = 0
                    ones = 0
                    for value in range(3):
                        if(imdata.__next__()[0] == 0):
                            zeroes += 1
                        elif(imdata.__next__()[0] == 1):
                            ones += 1
                    if(zeroes > ones):
                        if (pix[j]% 2 != 0): 
                            pix[j] -= 1
                    else:
                        continue
                
                 elif(pixel_flag == 1):
                    zeroes = 0
                    ones = 0
                    for value in range(3):
                        if(imdata.__next__()[0] == 0):
                            zeroes += 1
                        elif(imdata.__next__()[0] == 1):
                            ones += 1
                    if(zeroes < ones):
                        if (pix[j]% 2 != 0): 
                            pix[j] -= 1
                    else:
                        continue
                
                  
        # Eigh^th pixel of every set tells  
        # whether to stop or read further. 
        # 0 means keep reading; 1 means the 
        # message is over. 
        if (i == lendata - 1): 
            if (pix[-1] % 2 == 0): 
                pix[-1] -= 1
        else: 
            if (pix[-1] % 2 != 0): 
                pix[-1] -= 1
  
        pix = tuple(pix) 
        yield pix[0:3] 
        yield pix[3:6] 
        yield pix[6:9] 
  

          

def modPix(pix, data,pixel_flag): 
      
    datalist = genData(data) 
    lendata = len(datalist) 
    imdata = iter(pix) 
  
    for i in range(lendata): 
          
        # Extracting 3 pixels at a time 
        pix = [value for value in imdata.__next__()[:3] +
                                  imdata.__next__()[:3] +
                                  imdata.__next__()[:3]] 
                                      
        # Pixel value should be made  
        # odd for 1 and even for 0 
        for j in range(0, 8): 
            if (datalist[i][j]=='0') and (pix[j]% 2 != 0): 
                  
                if (pix[j]% 2 != 0): 
                    pix[j] -= 1
                      
            elif (datalist[i][j] == '1') and (pix[j] % 2 == 0): 
                pix[j] -= 1
                  
        # Eigh^th pixel of every set tells  
        # whether to stop ot read further. 
        # 0 means keep reading; 1 means the 
        # message is over. 
        if (i == lendata - 1): 
            if (pix[-1] % 2 == 0): 
                pix[-1] -= 1
        else: 
            if (pix[-1] % 2 != 0): 
                pix[-1] -= 1
  
        pix = tuple(pix) 
        yield pix[0:3] 
        yield pix[3:6] 
        yield pix[6:9] 
  
def encode_enc(newimg, data,pixel_flag): 
    w = newimg.size[0] 
    (x, y) = (0, 0) 
      
    for pixel in modPix(newimg.getdata(), data,pixel_flag): 
          
        # Putting modified pixels in the new image 
        newimg.putpixel((x, y), pixel) 
        if (x == w - 1): 
            x = 0
            y += 1
        else: 
            x += 1
              
# Encode data into image 
def encode(): 
    img = input("Enter image name(with extension): ") 
    image = Image.open(img, 'r') 
      
    data = input("Enter data to be encoded : ") 
    if (len(data) == 0): 
        raise ValueError('Data is empty') 
        
     #prompt user to choose where to encode : dark pixels or light pixels
    pixel_flag = input("Enter 0 for dark pixel encoding and 1 for light pixel encoding: ")
    while(int(pixel_flag) not in [0, 1]):
        print("sorry wrong input")
        pixel_flag = input("Enter 0 for dark pixel encoding and 1 for light pixel encoding: ")      
    newimg = image.copy() 
    encode_enc(newimg, data,pixel_flag) 
      
    new_img_name = input("Enter the name of new image(with extension): ") 
    newimg.save(new_img_name, str(new_img_name.split(".")[1].upper())) 
  
# Decode the data in the image 
def decode(): 
    img = input("Enter image name(with extension) :") 
    image = Image.open(img, 'r') 
    pixel_flag= input("Specify the type of encoding; Enter 0 for dark pixel encoding and 1 for light pixel encoding: ")
      
    binstrg=''
    data = '' 
    imgdata = iter(image.getdata()) 
      
    while (True): 
        pixels = [value for value in imgdata.__next__()[:3] +
                                  imgdata.__next__()[:3] +
                                  imgdata.__next__()[:3]] 
        # string of binary data 
        binstr = '' 
          
        for i in pixels[:8]: 
            if (i % 2 == 0):
                binstr +='0'
                if(pixel_flag == 0):
                    zeroes = 0
                    ones = 0
                    for value in range(3):
                        if(imdata.__next__()[0] == 0):
                            zeroes += 1
                        elif(imdata.__next__()[0] == 1):
                            ones += 1
                    if(zeroes > ones):
                        binstrg += '0'
                    else:
                        continue 
                
            else:
                binstr +='1'
                if(pixel_flag == 1):
                    zeroes = 0
                    ones = 0
                    for value in range(3):
                        if(imdata.__next__()[0] == 0):
                            zeroes += 1
                        elif(imdata.__next__()[0] == 1):
                            ones += 1
                    if(zeroes < ones):
                        binstrg += '1'
                    else:
                        continue        
        data += chr(int(binstr, 2)) 
        if (pixels[-1] % 2 != 0): 
            return data 
              
# Main Function         
def main(): 
    a = int(input(":: Welcome to Steganography ::\n"
                        "1. Encode\n 2. Decode\n")) 
    if (a == 1): 
        encode() 
          
    elif (a == 2): 
        print("Decoded word- " + decode()) 
    else: 
        raise Exception("Enter correct input") 
          
# Driver Code 
if __name__ == '__main__' : 
      
    # Calling main function 
    main() 
```
